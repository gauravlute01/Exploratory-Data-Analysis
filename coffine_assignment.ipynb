```python
import pandas as pd 
```

```python
# Q1 . Load the dataset and display the first 10 rows
df=pd.read_csv('coffee_survey.csv')
print(f'first 10 rows in coffee datasets:\n{df.head(10)}')
```

```python
# Q 2  How many rows and columns are present in the dataset?
rows,columns=df.shape
print(f'no of rows: {rows}')
print(f'no of columns: {columns}')
```

```python
# Q3  What are the data types of each column?
print(f'the data types of coffee :\n{df.dtypes}')
```

```python
# Q4 Identify the columns with missing data and their respective counts.
print('missing columns and their respective counts:\n')
print(df.isnull().sum())

print(df.isnull().sum()>0)
```

```python
# Q5 For columns with more than 20% missing values, suggest a method to handle them and implement it.
# more than 20 percent data is missing respctive columns this stage remove this columns 
# and lass than 20 percent missing values this stage fill using mean or previous value or 5- 10 percent missing to remove those rows  
missing_data=(df.isnull().sum())
missing_data[missing_data>0]
missing_percentage=(missing_data/rows)*100
columnn_drop=missing_percentage[missing_percentage>20].index
df_1=df.drop(columns=columnn_drop)

print(df_1)
```

```python
# remove the remaining missing value using rows removes 
new_df=df_1.dropna()
print(f'the final datasts\n:{new_df}')
```

```python
new_df.isnull().sum()
```

```python
# Generate summary statistics  for remaining features they are mostly categorical so i have choose describe data
new_df.describe()
```

```python
# Q3 a Generate summary statistics (mean, median, mode, etc.) for numerical columns (such as expertise, coffee_a_bitterness, and coffee_a_acidity).
num_columns=df.columns[df.dtypes=='float64']
for i in num_columns:
    print(f'mean of column{i} is: {df[i].mean()}')
    print(f'median of column{i} is: {df[i].median()}')
    print(f'mode of column{i} is: {df[i].mode()}')
    
```

```python
# b. Analyze bitterness preference across different age groups
bitterness_by_age = new_df.groupby('age')['coffee_a_bitterness'].mean()
print("\nBitterness preference (coffee_a_bitterness) across different age groups:")
print(bitterness_by_age)
```
